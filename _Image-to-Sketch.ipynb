# Converting Image to Pencil Sketch

## Beginner level (Task 4)

## Step 1 - Importing Libraries

```python

import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

## Step 2 - Loading Image

```python
img=cv2.imread("owl.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Owl")
plt.imshow(img)
plt.show()
```

## Step 3 - Converting to Grayscale Image

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Owl")
plt.imshow(gray_img,cmap="gray")
plt.show()
```

## Step 3 - Inverting the Image

```python
invert_img= 255-gray_img
plt.imshow(invert_img,cmap="gray")
plt.show()
```

## Step 4 - Blurring the Inverted image

```python
blur_img=cv2.GaussianBlur(invert_img,(21,21),0)
inverted_blur_img=255-blur_img
```

## Step 5 - Pencil Sketch Image

```python
pencil_sketch=cv2.divide(gray_img,inverted_blur_img,scale=256.0)
plt.title("Pencil Sketch of Owl")
plt.imshow(pencil_sketch,cmap="gray")
plt.show()
```
